## 1. Pancake Networks
Based on the original code, here is the modificaton to plot the pancakes networks for n - Pancakes , the graph show edges in both ways to travel across the network.

*The plots are done for $n$ between 2 to 5 due to the ammount of nodes(vertex) within each iteration, they become unable to be iddentified so it would be useless to plot them.*

```
import pydot
from collections import defaultdict 
from IPython.display import Image, display
```

```
def lstr(p):
        return ''.join([str(elem)  for elem in p])
```

```
def PancakeNet(s): 
#Modified for plotting networks
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="graph")
        
        queue = [] 
  
        queue.append(s) 
        visited[lstr(s)] = True
  
        while queue: 
  
            s = queue.pop(0) 
            graph[lstr(s)]=[]
            node = pydot.Node(lstr(s))
            G.add_node(node)
            
            for i in range(len(s)-1):
              d = s[i+1::-1] + s[i+2:]  
              if lstr(d) not in visited.keys():
                node = pydot.Node(lstr(d))
                G.add_node(node)
                queue.append(d) 
                graph[lstr(s)].append(lstr(d))
                visited[lstr(d)] = True
              if d in queue:
                edge = pydot.Edge(lstr(s),lstr(d))
                G.add_edge(edge)
              
        im = Image(G.create_png())
        display(im)
```

### Plotting whit n = 2 pancakes

```
p2 = [2,1]
PancakeNet(p2)
```

### Plotting with n = 3 pancakes

```
p3 = [2,3,1]
PancakeNet(p3)
```

### Plotting with n = 4 pancakes.

```
p4 = [2,3,4,1]
PancakeNet(p4) 
```

### Plotting with n = 5 Pancakes:

```
p5 = [1,3,5,2,4]
PancakeNet(p5)
```

```
def pnNumbers(s):  

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = set()
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append((s,0)) 
        visited.add(lstr(s))
        p = 0
        while queue: 
            # Dequeue a vertex from queue 
            e = queue.pop(0)
            s = e[0]
            l = e[1]
            p = max(l,p)
            for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            for i in range(len(s)-1):
              d = s[i+1::-1] + s[i+2:]  
              if (d != s) and not(lstr(d) in visited):
                #node = pydot.Node(lstr(d))
                queue.append((d,l+1)) 
                visited.add(lstr(d))

        return p
```

##**Pancake sorting:** Two Spatulas

Modify the code to plot the pancakes trees using two spatulas


```
def Pancakes2S(s): 
#modified for plotting trees but two spatulas at a time

        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        queue = [] 
  
        queue.append(s) 
        visited[lstr(s)] = True
  
        while queue: 
   
            s = queue.pop(0) 
            graph[lstr(s)]=[]

            flag = False
            for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]
                if(d == sorted(d)):
                     flag = True
                     
            node = pydot.Node(lstr(s))
            G.add_node(node)
            
            if(s != sorted(s)):
              for i in range(len(s)):
                for j in range(i+1, len(s)):
                  d = s[:i] + s[j:i:-1] + [s[i]] + s[j+1:]  
                  if (d != s) and not(lstr(d) in visited.keys()):
                    node = pydot.Node(lstr(d))
                    G.add_node(node)
                    edge = pydot.Edge(lstr(s),lstr(d))
                    G.add_edge(edge)
                    graph[lstr(s)].append(lstr(d))
                    queue.append(d) 
                    visited[lstr(d)] = True 
              
        im = Image(G.create_png())
        display(im)
```
