## SageMath’s Integrals

Based on the documentation provided for SageMath, it has functions made to evaluate defined and undifined integrals, and integral transforms such as Fourier and Meillin.

In this case we'll evaluate integrals of one dimensional an three dimensional functions that are product of a polynomial, $log$ , $sin$, and $e$.

### Polynomial: $ x^3 + 2x^{5} -17x^{4}$

we will do the undefined integral, and the defined integral evaluated between $0$ and $7$:

$$ \int{x^3 + 2x^{5} -17x^{4}} dx \;; \ \int_{0}^{7} x^3 + 2x^{5} -17x^{4} dx $$

Solutions:

$$ \int{x^3 + 2x^{5} -17x^{4}}  dx = \frac{x^4}{4} + \frac{x^6}{3} - \frac{17x^5}{5} + C$$

And:

$$ \int_{0}^{7} x^3 + 2x^{5} -17x^{4} dx  dx = \frac{x^4}{4} + \frac{x^6}{3} - \frac{17x^5}{5}  = -\frac{1039633}{60} \approx -17327.21...$$

Sin título4.png

```python
from sage.symbolic.integration.integral import definite_integral
from sage.symbolic.integration.integral import indefinite_integral
x=var('x')
f1 = x**3 + 2*(x**5) - 17*(x**4)
i1=indefinite_integral(f1,x)
print("Undefined Integral: ",i1)
di1=definite_integral(f1,x,0,7)
print("Defined Integral between 0 and 7: ",di1)
```

#### Integral using Montecarlo
We can estimate this integral using a standard Monte Carlo method, where we use the fact that the expectation of a random variable is related to its integral

$$\mathbb{E}(f(x)) = \int_I f(x) dx $$

We will sample a large number N of points in I and calculate their average, and multiply by the range over which we are integrating

```python
from numpy import *
N = 1_000_000
accum = 0
for i in range(N):
    R = random.uniform(0, 7)
    accum += R**3 + 2*(R**5) - 17*(R**4)
volume = 7 - 0
print("Standard Monte Carlo result:",volume * accum / float(N))
```

### Logarithm: $\frac{Ln(x^2)}{x}$

we will do the undefined integral considering Natural Logarithm, and the defined integral evaluated between $4$ and $6$:

$$ \int{ \frac{Ln(x^2)}{x}} dx \;; \ \int_{4}^{6} \frac{Ln(x^2)}{x} dx $$

Solutions:

$$ \int{\frac{Ln(x^2)}{x}}  dx = \frac{2-Ln(x^2)}{x^2} + C$$

And:

$$ \int_{4}^{6}\frac{Ln(x^2)}{x} dx = \frac{2-Ln(x^2)}{x^2} = \frac{-ln(16)^2}{4} + \frac{ln(36)^2}{4} \approx 1.2886... $$

```python
from sage.symbolic.integration.integral import definite_integral
from sage.symbolic.integration.integral import indefinite_integral
x=var('x')
f2 = (ln(x^2))/x
i2=indefinite_integral(f2,x)
print("Undefined Integral: ",i2)
di2=definite_integral(f2,x,4,6)
print("Defined Integral between 4 and 6: ",di2)
res = float(di2)
print("value",res)
```

#### Integral using Montecarlo
We can estimate this integral using a standard Monte Carlo method, where we use the fact that the expectation of a random variable is related to its integral

$$\mathbb{E}(f(x)) = \int_I f(x) dx $$

We will sample a large number N of points in I and calculate their average, and multiply by the range over which we are integrating.

```python
from numpy import *
N = 10000
accum = 0
for i in range(N):
    r = random.uniform(4, 6)
    accum += ln(r**2)/r
volume = 6 - 4
print("Standard Monte Carlo result:",volume * accum / float(N))
```

### Sine: $x\cdot sin(x)$

we will do the undefined integral considering Natural Logarithm, and the defined integral evaluated between $-1$ and $3$:

$$ \int{x\cdot sin(x)} dx \;; \ \int_{-1}^{3} x\cdot sin(x) dx $$

$$ \int{x\cdot sin(x)}  dx = sin(x) - xcos(x) + C$$

And:


$$ \int_{-1}^{3} {x\cdot sin(x)}  dx = sin(x) - xcos(x) = -cos(1) + sin(3) + sin(1) - 3cos(3) \approx 3.4123... $$
